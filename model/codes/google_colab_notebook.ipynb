```
from ultralytics import YOLO

model = YOLO('yolo11s-cls.pt') # or 'yolo11m-cls.pt' if GPU allows

model.train(

data='/content/tomato_h-bs-lb_full_sl.v1i.folder',

epochs=100,

imgsz=224, # better for classification

batch=32,  # reduce to 16 if memory error

patience=10,

optimizer='AdamW',

lr0=0.001,

dropout=0.2,

augment=True,

freeze=10,# helps generalization

workers=4,

project='/content/drive/MyDrive/my_training_results',

)
```
